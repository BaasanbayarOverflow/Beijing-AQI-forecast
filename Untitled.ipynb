```python
dataframe = pd.read_csv(r'.\ultimate.csv')
dataframe = dataframe.set_index(pd.DatetimeIndex(dataframe['dtime']))
date_index_series = dataframe['dtime']
dataframe = dataframe.drop(columns=['dtime'])
timeseries = TimeSeries.from_dataframe(dataframe)
dataframe.head()
```

```python
START_DATE = pd.Timestamp('20160101')
DATAFRAME_METRICS = pd.DataFrame()

def evaluateModel(timeseries_, prediction_, model_name:str, column_name:str):  
    mae_ = mae(timeseries_, prediction_)
    rmse_ = rmse(timeseries_, prediction_)
    mape_ = mape(timeseries_, prediction_)
    smape_ = smape(timeseries_, prediction_)
    r2_score_ = r2_score(timeseries_, prediction_)
    
    dict_ = {
        'MAE': mae_, 'RMSE': rmse_,
        'MAPE': mape_, 'SMAPE': smape_,
        'R2': r2_score_, 'Model_name': model_name, 
        'Col_name': column_name, 
    }
    dataframe = pd.DataFrame(dict_, index=[model_name])
    return (dataframe.round(decimals=2))
```

```python
model = LinearRegressionModel(lags=12)

columns = dataframe.columns
prediction_list = []
for col in columns:
    prediction = model.historical_forecasts(timeseries[col], start=START_DATE, forecast_horizon=12, verbose=True)
    prediction_list.append(prediction)
    DATAFRAME_METRICS = DATAFRAME_METRICS.append(evaluateModel(timeseries[col], prediction, 'LinearRegression', col))
```

```python
i = 1
fig = plt.figure(figsize=(16, 20))
for col in columns:
    series = prediction_list[i - 1].pd_dataframe() 
    
    ax = fig.add_subplot(4, 2, i)
    ax.grid(False)
    ax.set_title(col, size=14)
    ax.plot(dataframe[col][-500:], label='Actual Values', lw=1.2)
    ax.plot(series, label='Forecast', color='orange', lw=1.5)
    ax.legend()
    i += 1
```

```python
TRAIN_SIZE = 1047
TEST_SIZE = 414
LOOK_BACK = 60

def createDataset(dataset, look_back:int):
    x, y = [], []
    for i in range(len(dataset) - look_back - 1):
        temp = dataset[i:(i+look_back), 0]
        x.append(temp)
        y.append(dataset[i + look_back, 0])
    return np.array(x), np.array(y)

def processDataset(column_name:str):
    dataset = dataframe[column_name].values.astype(np.float32)
    dataset = dataset.reshape(-1, 1)
    norm = MinMaxScaler(feature_range=(0, 1))
    dataset = norm.fit_transform(dataset)
    train, test = dataset[:TRAIN_SIZE, :], dataset[TRAIN_SIZE:len(dataset), :]
    
    X_train, y_train = createDataset(train, LOOK_BACK)
    X_test, y_test = createDataset(test, LOOK_BACK)

    X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
    X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
    return X_train, y_train, X_test, y_test
```

```python
model = Sequential()
model.add(LSTM(100, input_shape=(1, LOOK_BACK)))
model.add(Dense(50))
model.add(Dense(1))
model.compile(loss=mean_squared_error, optimizer='adam')
```

```python
all_data = []
all_history = []
columns = dataframe.columns
for col in tqdm(columns):
    X_train, y_train, X_test, y_test = processDataset(col)
    history = model.fit(X_train, y_train, epochs=100, batch_size=50, 
                        validation_data=(X_test, y_test), verbose=0, shuffle=False)
    all_history.append(history)
```
