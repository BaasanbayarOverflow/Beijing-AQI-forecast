```python
def evaluateModel(timeseries_, prediction_, model_name:str, column_name:str):  
    length = len(prediction_)
    actual = timeseries_[-length:].pd_series().values
    forecast = prediction_.pd_series().values
    mae_ = mae(actual, forecast)
    rmse_ = sqrt(mse(actual, forecast))
    mape_ = mape(actual, forecast)
    r2_score_ = r2_score(actual, forecast)
    
    dict_ = {
        'MAE': mae_, 'RMSE': rmse_,
        'MAPE': mape_, 'R2': r2_score_, 
        'Model_name': model_name, 
        'Col_name': column_name, 
    }
    dataframe = pd.DataFrame(dict_, index=[model_name])
    return (dataframe.round(decimals=2))
```

<h4>LSTM model</h4>

```python
model = Sequential()
model.add(LSTM(200, input_shape=(1, LOOK_BACK), return_sequences=True))
model.add(LSTM(100, input_shape=(1, LOOK_BACK)))
model.add(Dense(50))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss=mean_squared_error, optimizer='adam')
```

```python
def createDataset(dataset, look_back:int):
    x, y = [], []
    for i in range(len(dataset) - look_back - 1):
        temp = dataset[i:(i + look_back), 0]
        x.append(temp)
        y.append(dataset[i + look_back, 0])
    return np.array(x), np.array(y)
```

```python
columns = dataframe.columns

test_all = []
forecast_all = []
histories_all = dict()
for col in tqdm(columns):
    dataset = dataframe[col].values.astype(np.float32)
    dataset = dataset.reshape(-1, 1)
    
    normalization = MinMaxScaler(feature_range=(0, 1))
    dataset_normalized = normalization.fit_transform(dataset)
    
    train = dataset_normalized[:TRAIN_SIZE, :] 
    test = dataset_normalized[TRAIN_SIZE:len(dataset), :]

    X_train, y_train = createDataset(train, LOOK_BACK)
    X_test, y_test = createDataset(test, LOOK_BACK)

    X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
    X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
    
    # Training process
    history = model.fit(X_train, y_train, epochs=100, batch_size=50, 
                        validation_data=(X_test, y_test), verbose=0, shuffle=False)
    histories_all[col] = history
    
    file_name = col.replace('.', '_') if ('.' in col) else col
    file_name = 'LSTM_' + file_name + '.h5'
    file_path = r'.\models\\' + file_name
    
    if (os.path.exists(file_path)):
        os.remove(file_path)
    
    try:
        tf.keras.models.save_model(model, filepath = (file_path))
    except:
        print('Cannot save model!')
                               
    # Forecasting and Validating model 
    forecast_on_testing_data = model.predict(X_test)
    forecast_inv_test = normalization.inverse_transform(forecast_on_testing_data)
    y_test_inv = normalization.inverse_transform([y_test])
    
    test_all.append(y_test_inv[0])
    forecast_all.append(forecast_inv_test)
    
    test_mae = mae(y_test_inv[0], forecast_inv_test[:, 0])
    test_rmse = sqrt(mse(y_test_inv[0], forecast_inv_test[:, 0]))
    test_mape = mape(y_test_inv[0], forecast_inv_test[:, 0])
    test_r2_score = r2_score(y_test_inv[0], forecast_inv_test[:, 0])
    
    
    dict_ = {
        'MAE': test_mae, 'RMSE': test_rmse,
        'MAPE': test_mape, 'R2': test_r2_score, 
        'Model_name': 'LSTM', 'Col_name': col, 
    }
    
    temp = pd.DataFrame(dict_, index=['LSTM'])
    DATAFRAME_METRICS = DATAFRAME_METRICS.append(temp.round(decimals=2))
```

```python
index = 1
fig = plt.figure(figsize=(18, 13))
for col in columns:
    ax = fig.add_subplot(3, 3, index)
    ax.set_title(col, size=14)
    ax.plot(forecast_all[index - 1])
    ax.plot(test_all[index - 1], color='orange')
    index += 1
    
fig.subplots_adjust(left=0.1, bottom=0.1, right=1., top=1., wspace=0.2, hspace=0.3)
```
